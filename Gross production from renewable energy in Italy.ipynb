# Gross production of electricity from renewable sources



## Annual changes in renewable souces


data source : http://dati.istat.it/Index.aspx?DataSetCode=DCCV_PRODENERG


```python
import csv 
import matplotlib.pyplot as plt
import numpy as np

Idroyears = []
Eolicyears = []
Pvyears = []
Geoyears = []
Bioyears = []

years = ["2013","2014","2015","2016","2017"]
label = ["idroelettric","eolic","photovoltaic","geotermical","bio and waste"]

for year in years:
    with open("DCCV_PRODENERG_31052023221602525.csv", newline ="") as csvfile:
        reader = csv.DictReader(csvfile, delimiter=";")
        for row in reader:
        #print(row)
            if row["TIME"] == year:
                if row["TIPO_FONTERINN"] == "1":
                    Idroyears.append(float(row["Value"]))
                if row["TIPO_FONTERINN"] == "2":
                    Eolicyears.append(float(row["Value"]))
                if row["TIPO_FONTERINN"] == "3":
                    Pvyears.append(float(row["Value"]))
                if row["TIPO_FONTERINN"] == "4":
                    Geoyears.append(float(row["Value"]))
                if row["TIPO_FONTERINN"] == "5":
                    Bioyears.append(float(row["Value"]))

Energy = [Idroyears,Eolicyears,Pvyears,Geoyears,Bioyears]

n = 0
for E in Energy:
    fig, ax = plt.subplots()
    ax.bar(years, E, color = ("red","green","blue","orange","purple"))
    ax.set_ylabel("Million of KWh")
    ax.set_title(label[n])
    n = n+1
                
```

## Predominant annual source of energy 

```python
i = 0
an_year = []

for E in Energy:
    for T in Energy:
        an_year.append(T[i])
    fig, ax = plt.subplots()
    ax.pie(x = an_year,autopct='%1.2f%%',wedgeprops={"linewidth": 1, "edgecolor": "white"}, labels = label, explode=[0.05, 0.05, 0.05, 0.05, 0.05])
    ax.set_title(years[i])
    
    i = i +1
    an_year = []
```

## Growth in 5 years


    

```python
n = 0
diffp = []
for T in Energy:
    diff = T[4]-T[0]
    diffp.append(diff/(T[0]/100))
    print(label[n],round(diffp[n],2),"%")
    n=n+1
fig,ax = plt.subplots()
ax.bar(label, diffp)
ax.set_title("% growth in 5 years")
```

In the graph above we can see that the most growth in energy production was dominated by eolic sector, followed by bio, photovoltaic and geotermical resouces. In the last position, with a negative resoult we can see the Idroelettric with -31.41% in 5 years.
