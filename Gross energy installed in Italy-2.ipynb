# Gross production of electricity from renewable sources



## Predominant 


```python
import csv 
import matplotlib.pyplot as plt

time = []
Value = []
Type = []
Installed13 = []
Type13 = []
Idroyears = []
Eolicyears = []
Pvyears = []
Geoyears = []
Bioyears = []
IdroTotInstall = 0
EolicaTotInstall = 0
PVTotInstall = 0
GEOTotInstall = 0
BIOTotInstall = 0

years = ["2013","2014","2015","2016","2017"]
Energy = ["idroelettrica","eolica","fotovoltaica","geotermica","biomasse e rifiuti"]

for year in years:
    with open("DCCV_PRODENERG_31052023221602525.csv", newline ="") as csvfile:
        reader = csv.DictReader(csvfile, delimiter=";")
        for row in reader:
        #print(row)
            if row["TIME"] == year:
                if row["Tipologia della fonte rinnovabile"] == "idroelettrica":
                    IdroTotInstall = IdroTotInstall + float(row["Value"])
                elif row["Tipologia della fonte rinnovabile"] == "eolica":
                    EolicaTotInstall = EolicaTotInstall + float(row["Value"])
                elif row["Tipologia della fonte rinnovabile"] == "fotovoltaica":
                    PVTotInstall = PVTotInstall + float(row["Value"])
                elif row["Tipologia della fonte rinnovabile"] == "geotermica":
                    GEOTotInstall = GEOTotInstall + float(row["Value"])
                elif row["Tipologia della fonte rinnovabile"] == "biomasse e rifiuti":
                    BIOTotInstall = BIOTotInstall + float(row["Value"])
        #print (year, " -- ",round(IdroTotInstall,2),round(BIOTotInstall,2),round(GEOTotInstall,2),round(PVTotInstall,2),round(EolicaTotInstall,2))
        power = [round(IdroTotInstall,2),round(BIOTotInstall,2),round(GEOTotInstall,2),round(PVTotInstall,2),round(EolicaTotInstall,2)]
        Idroyears.append(round(IdroTotInstall,2))
        Eolicyears.append(round(EolicaTotInstall,2))
        Pvyears.append(round(PVTotInstall,2))
        Geoyears.append(round(GEOTotInstall,2))
        Bioyears.append(round(BIOTotInstall,2))
        
        fig, ax = plt.subplots()
        ax.bar(Energy, power)
        plt.xticks(rotation = 15)
        ax.set_ylabel("Million of KWh")
        ax.set_title(year)    
tip = [Idroyears,Eolicyears,Pvyears,Geoyears,Bioyears]
```

## Annual growth 

```python
import numpy as np
n = 0
for T in Energy:
    fig, ax = plt.subplots()
    ax.bar(years, tip[n])
    ax.set_ylabel("Million of KWh")
    ax.set_title(T)  
    n = n+1
```

## Grow in 5 years


    

```python
n = 0
per = []
for t in tip:
    d = t[4]-t[0]
    pp = t[0]/100
    r = d/pp
    per.append(round(r,2))
    print(Energy[n],"+"+str(round(r,2)),"%")
    n = n+1
    #Idro, Eolic, Pv, Geo, Bio
    

fig, ax = plt.subplots()
ax.bar(Energy, per)
plt.xticks(rotation = 15)
ax.set_ylabel("percentage (%)")
ax.set_title("% growth in 5 years")  
n = n+1
```
